```python
import pyautogui
import time
import ctypes
```

# Using `pyautogui` to refresh map gen on editor `.rms` save

```python
pyautogui.position()
```

```python
# click at cursor location
pyautogui.click(pyautogui.position())
```

```python
pos = pyautogui.position()

# move to button
ctypes.windll.user32.SetCursorPos(-2000,1280)
pyautogui.click(pyautogui.position())

# escape editor
pyautogui.keyDown('alt')
time.sleep(.05)
pyautogui.press('tab')
time.sleep(.05)
pyautogui.keyUp('alt')

# return cursor to original position
ctypes.windll.user32.SetCursorPos(pos[0], pos[1])
pyautogui.click(pyautogui.position())
```
